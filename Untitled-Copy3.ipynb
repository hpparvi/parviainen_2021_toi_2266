```python
from ldtk import LDPSetCreator, sdss_g, sdss_r, sdss_i, sdss_z, tess
```

```python
filters = [tess, sdss_g, sdss_r, sdss_i, sdss_z]
```

```python
sc = LDPSetCreator((3200, 100), (5, 0.25), (0.08, 0.08), filters, dataset='vis-lowres')
```

```python
ps = sc.create_profiles()
```

```python
ps.coeffs_tq(do_mc=True)
```

```python
lpf = LPF('TOI-2266b-riz', passbands=('g', 'r', 'i', 'z_s'), downsample=60)
lpf.set_prior('tc', 'NP', 2459255.6944, 0.005)
lpf.set_prior('p', 'NP', 2.32632, 5.0e-5)
lpf.set_prior('rho', 'UP', 10, 30)
lpf.set_prior('k2_app', 'UP', 0.02**2, 0.08**2)
lpf.set_prior('k2_true', 'UP', 0.02**2, 0.95**2)
lpf.set_prior('k2_app_tess', 'UP', 0.02**2, 0.08**2)
```

```python
lpf.set_prior('q1_tess', 'NP', 0.37, 0.03)
lpf.set_prior('q2_tess', 'NP', 0.25, 0.02)
lpf.set_prior('q1_g', 'NP', 0.74, 0.03)
lpf.set_prior('q2_g', 'NP', 0.33, 0.02)
lpf.set_prior('q1_r', 'NP', 0.68, 0.03)
lpf.set_prior('q2_r', 'NP', 0.36, 0.02)
lpf.set_prior('q1_i', 'NP', 0.48, 0.03)
lpf.set_prior('q2_i', 'NP', 0.26, 0.02)
lpf.set_prior('q1_z_s', 'NP', 0.36, 0.03)
lpf.set_prior('q2_z_s', 'NP', 0.23, 0.02)
```

```python
lpf.ps
```

```python
lpf.optimize_global(niter=500, npop=600)
```

```python
lpf.optimize_global(niter=5000, npop=600)
```
