```python
from pytransit.lpf.tesslpf import downsample_time
from pytransit.utils.misc import fold
```

```python
pv = lpf.de.minimum_location
```

```python
bl = lpf.baseline(pv)
fo = lpf.ofluxa / bl
fm = lpf.transit_model(pv)
```

```python
variances = ((10**pv[lpf._sl_wn])[lpf.lcids])**2
```

```python
from numba import njit

@njit
def downsample_time_weighted(time, vals, variances, inttime=1.):
    duration = time.max() - time.min()
    nbins = int(ceil(duration / inttime))
    bins = arange(nbins)
    edges = time[0] + bins * inttime
    bids = digitize(time, edges) - 1
    bt, bv, be = full(nbins, nan), zeros(nbins), zeros(nbins)
    for i, bid in enumerate(bins):
        bmask = bid == bids
        if bmask.sum() > 0:
            n = (1./variances[bmask]).sum()
            bt[i] = (time[bmask] / variances[bmask]).sum() / n
            bv[i] = (vals[bmask] / variances[bmask]).sum() / n
            if bmask.sum() > 2:
                be[i] = sqrt(1. / (1. / variances[bmask]).sum())
            else:
                be[i] = nan
    m = isfinite(be)
    return bt[m], bv[m], be[m]
```

```python
fig, axs = subplots(1, 4, figsize=(13,3), sharex='all', sharey='all')

for ipb in range(4):
    ax = axs.flat[ipb]
    m = lpf.pbids == ipb

    ttime, tfo, tfm = [], [], []
    for i,sl in enumerate(lpf.lcslices):
        if m[i]:
            ttime.append(lpf.times[i])
            tfo.append(fo[sl])
            tfm.append(fm[sl])
    ttime, tfo, tfm = concatenate(ttime), concatenate(tfo), concatenate(tfm)

    phase = (fold(ttime, pv[1], pv[0], 0.5) - 0.5) * pv[1]
    sids = argsort(phase)

    pb, fb, eb = downsample_time_weighted(phase[sids], tfo[sids], variances[sids], 0.003)

    ax.errorbar(24*pb, fb, eb, fmt='.')
    ax.plot(24*phase[sids], tfm[sids], 'k')
setp(ax, xlim=(-1.1, 1.1))
fig.tight_layout()
```
