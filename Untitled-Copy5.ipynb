```python
from ldtk import LDPSetCreator, sdss_g, sdss_r, sdss_i, sdss_z, tess
```

```python
filters = [tess, sdss_g, sdss_r, sdss_i, sdss_z]
```

```python
sc = LDPSetCreator((3200, 100), (5, 0.25), (0.08, 0.08), filters, dataset='vis-lowres')
```

```python
ps = sc.create_profiles(1000)
```

```python
ps.resample_linear_z(200)
```

```python
ps.reset_sampling()
```

```python
aa, bb = ps.coeffs_p2mp(do_mc=True)
```

```python
from pytransit.models.numba.ldmodels import ld_power_2_pm, evaluate_ld
```

```python
bb
```

```python
fig, ax = subplots(figsize=(13,6))
plot(ps._mu, evaluate_ld(ld_power_2_pm, ps._mu, aa[newaxis,:,:])[0].T)
plot(ps._mu, ps.profile_averages.T, '.');
fig.tight_layout()
```

```python
lpf = LPF('TOI-2266b-b', passbands=('g', 'r', 'i', 'z_s'), downsample=60)
lpf.set_prior('tc', 'NP', 2459255.6944, 0.005)
lpf.set_prior('p', 'NP', 2.32632, 5.0e-5)
lpf.set_prior('rho', 'UP', 10, 40)
lpf.set_prior('teff_c', 'NP', 3200, 160)
lpf.set_prior('teff_h', 'UP', 2500, 12000)
lpf.set_prior('k2_app', 'UP', 0.02**2, 0.08**2)
lpf.set_prior('k2_true', 'UP', 0.02**2, 0.95**2)
lpf.set_prior('k2_app_tess', 'UP', 0.02**2, 0.08**2)
```

```python
lpf.set_prior('q1_tess', 'NP', 0.40, 0.03)
lpf.set_prior('q2_tess', 'NP', 0.23, 0.02)
lpf.set_prior('q1_g', 'NP', 0.82, 0.03)
lpf.set_prior('q2_g', 'NP', 0.31, 0.02)
lpf.set_prior('q1_r', 'NP', 0.72, 0.03)
lpf.set_prior('q2_r', 'NP', 0.34, 0.02)
lpf.set_prior('q1_i', 'NP', 0.53, 0.03)
lpf.set_prior('q2_i', 'NP', 0.24, 0.02)
lpf.set_prior('q1_z_s', 'NP', 0.43, 0.03)
lpf.set_prior('q2_z_s', 'NP', 0.19, 0.02)
```
